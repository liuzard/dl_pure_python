##### 1、第一种方式实现感知机

```python

def AND(x1,x2):
    """与门"""
    w1,w2=0.5,0.5
    theta=0.7
    result=w1*x1+w2*x2
    if result >theta:
        return 1
    else:
        return 0

def NAND(x1,x2):
    """与非门"""
    w1,w2=-0.5,-0.5
    theta=-0.7
    result=w1*x1+w2*x2
    if result>theta:
        return 1
    else:
        return 0
    
    
def OR(x1,x2):
    """或门"""
    w1,w2=0.5,0.5
    theta=0.3
    result=w1*x1+w2*x2
    if result>theta:
        return 1
    else:
        return 0

def XOR(x1,x2):
    """异或门"""
    s1=OR(x1,x2)
    s2=NAND(x1,x2)
    return AND(s1,s2)
    
    
print("与门测试:")
print("\tAND(1,1)-->",AND(1,1))
print("\tAND(1,0)-->",AND(1,0))
print("\tAND(0,1)-->",AND(0,1))
print("\tAND(0,0)-->",AND(0,0))

print("与非门测试:")
print("\tNAND(1,1)-->",NAND(1,1))
print("\tNAND(1,0)-->",NAND(1,0))
print("\tNAND(0,1)-->",NAND(0,1))
print("\tNAND(0,0)-->",NAND(0,0))

print("或门测试:")
print("\tOR(1,1)-->",OR(1,1))
print("\tOR(1,0)-->",OR(1,0))
print("\tOR(0,1)-->",OR(0,1))
print("\tOR(0,0)-->",OR(0,0))

print("异或门测试:")
print("\tXOR(1,1)-->",XOR(1,1))
print("\tXOR(1,0)-->",XOR(1,0))
print("\tXOR(0,1)-->",XOR(0,1))
print("\tXOR(0,0)-->",XOR(0,0))
```

##### 2、第2种方式实现感知机

```python
import numpy as np


def AND(x1,x2):
    """与门"""
    x=np.array([x1,x2])
    w=np.array([0.5,0.5])
    b=-0.7
    result=np.sum(w*x)+b
    if result>0:
        return 1
    else:
        return 0
    

def NAND(x1,x2):
    """与非门"""
    x=np.array([x1,x2])
    w=np.array([-0.5,-0.5])
    b=0.7
    result=np.sum(w*x)+b
    if result>0:
        return 1
    else:
        return 0
    
    
def OR(x1,x2):
    """或门"""
    x=np.array([x1,x2])
    w=np.array([0.5,0.5])
    b=-0.3
    result=np.sum(w*x)+b
    if result>0:
        return 1
    else:
        return 0

def XOR(x1,x2):
    """异或门"""
    s1=OR(x1,x2)
    s2=NAND(x1,x2)
    return AND(s1,s2)
    
    
print("与门测试:")
print("\tAND(1,1)-->",AND(1,1))
print("\tAND(1,0)-->",AND(1,0))
print("\tAND(0,1)-->",AND(0,1))
print("\tAND(0,0)-->",AND(0,0))

print("与非门测试:")
print("\tNAND(1,1)-->",NAND(1,1))
print("\tNAND(1,0)-->",NAND(1,0))
print("\tNAND(0,1)-->",NAND(0,1))
print("\tNAND(0,0)-->",NAND(0,0))

print("或门测试:")
print("\tOR(1,1)-->",OR(1,1))
print("\tOR(1,0)-->",OR(1,0))
print("\tOR(0,1)-->",OR(0,1))
print("\tOR(0,0)-->",OR(0,0))

print("异或门测试:")
print("\tXOR(1,1)-->",XOR(1,1))
print("\tXOR(1,0)-->",XOR(1,0))
print("\tXOR(0,1)-->",XOR(0,1))
print("\tXOR(0,0)-->",XOR(0,0))
```
